```python
%load_ext autoreload
%autoreload 2
import config as cfg
from sca import plots, helpers as h
```

```python
import numpy as np
X, y, pts, ks = h.load_data(cfg.DATA_DIR / 'ascadv_clean.h5')
best_feats_idx = np.load('data/rf_rfe_5_best.npy')

X_best = X[:, best_feats_idx[:5]]
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB

score = np.mean(h.cv(GaussianNB(), X_best, y, pts, ks, seed=cfg.SEED))
print(score)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=1, solver='eigen', shrinkage='auto')
score = np.mean(h.cv(lda, X_best, y, pts, ks, seed=cfg.SEED))
print(score)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

qda = QDA(solver='eigen')
score = np.mean(h.cv(qda, X_best, y, pts, ks, seed=cfg.SEED))
print(score)
```

```python
qda_s = QDA(solver='eigen', shrinkage='auto')
score = np.mean(h.cv(qda_s, X_best, y, pts, ks, seed=cfg.SEED))
logging.info(f"Mean CI: {score:2e}")
```

```python
import logging
params = [0.0, 0.01, 0.05, 0.1, 0.2, 0.5, 0.8, 0.95]

for reg in params:
    model = QDA(solver='eigen', shrinkage='auto', reg_param=reg)
    score = np.mean(h.cv(model, X_best, y, pts, ks, seed=cfg.SEED, verbose=False))
    logging.info(f"PI ({reg=}): {score:.2e}")
```

```python
from sklearn.preprocessing import PowerTransformer

pl = make_pipeline(
    PowerTransformer(method='yeo-johnson'),
    QDA(solver='eigen', shrinkage='auto')
)
score = np.mean(h.cv(pl, X_best, y, pts, ks, seed=cfg.SEED))
logging.info(f"Mean CI: {score:2e}")
```
